```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
topic_url = "https://github.com/topics"
```

```python
response = requests.get(topic_url)
```

```python
response.status_code
```

```python
len(response.text)
```

```python
page_contents = response.text
```

```python
page_contents
```

```python
doc = BeautifulSoup(page_contents,'html.parser')
```

```python
selection_class1 = 'f3 lh-condensed mb-0 mt-1 Link--primary'
topic_title_tags = doc.find_all('p',{'class' : selection_class1})
```

```python
topic_title_tags
```

```python
desc_selectors = 'f5 color-fg-muted mb-0 mt-1'
topic_desc_tags = doc.find_all('p',{'class':desc_selectors})
```

```python
topic_desc_tags
```

```python
topic_title_tag0=topic_title_tags[0]
```

```python
div_tag = topic_title_tag0.parent
```

```python
div_tag
```

```python
topic_link_tags = doc.find_all('a',{'class': "no-underline flex-1 d-flex flex-column"})
```

```python
len(topic_link_tags)
```

```python
topic_link_tags[0]['href']
```

```python
topic0_url = "https://github.com" + topic_link_tags[0]['href']
print(topic0_url)
```

```python
topic_titles = []
for tag in topic_title_tags:
    topic_titles.append(tag.text)
print(topic_titles)
```

```python
topic_urls = [] 
base_url = "https://github.com"
for tag in topic_link_tags:
    topic_urls.append(base_url + tag['href'])
topic_urls  
```

```python
topics_dict = {
    "title": topic_titles, 
    "description": topic_desc_tags, 
    "url":topic_urls
}
```

```python
topics_dict
```

```python
topics_df = pd.DataFrame(topics_dict)
```
